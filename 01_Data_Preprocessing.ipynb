<h2>8.Handling 'unknown' values</h2>
<p> Categorical columns may contain 'unknown' values.</p>
<p> These are handled by:</p>
<p> - Replacing with mode</p>
<p> - Ensuring data consistency</p>
<p> This prevents loss of useful data.</p>

```python
cat_cols = df.select_dtypes(include=['object','category']).columns
for col in cat_cols:
    unk = (df[col] == 'unknown').sum()
    if unk > 0:
        pct = round( unk/ len(df) * 100,1)
        print(f'{col:12s}: {unk:5d} unknowns ({pct}%)')
```

<h2>9.Data Cleaning </h2>
<p> For columns like 'job' ,'unknown' values are replaced using mode.</p>
<p> Helps in maintain realistic data distribution</p>

```python
job_mode = df[df['job'] != 'unknown']['job'].mode()[0]
df['job'] = df['job'].replace('unknown', job_mode)
print(f'job mode used: {job_mode}  | remaining unknowns: {(df["job"]=="unknown").sum()}')

edu_mode = df[df['education'] != 'unknown']['education'].mode()[0]
df['education'] = df['education'].replace('unknown', edu_mode)
print(f'education mode used: {edu_mode}  | remaining unknowns: {(df["education"]=="unknown").sum()}')

print('contact & poutcome unknowns kept as a valid category.')
```

<h2>10.Outlier Treatment(IQR Method)</h2>
<p>Outliers are handled using the IQR (InterQuartile Range) method:</p>
<p>-values outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR] are capped</p>
<p>Thi prevents extreme vales from skewing the model.</p>


```python
def cap_iqr(df,col):
    Q1, Q3= df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    n= df[(df[col]<lower) | (df[col] > upper)].shape[0]
    df[col] = df[col].clip(lower,upper)
    print(f" {col:12s} [{lower:8.1f},{upper:8.1f}] | {n} capped") 
    return df

print("capping outliers...\n") 
for col in ["age","balance","duration" ,"campaign"]:
    df = cap_iqr(df, col)
```

<h2>11.Final Cleaning of 'unknown'</h2>
<p> Final Cleaning of remainng 'unknown' values</p>
<p>A Final pass is done to ensure no 'unknown' values remain in categorical Clean data = Better model performance</p>


```python
for col in df.select_dtypes(include="object").columns:
    u = (df[col] =="unknown").sum()
    if u>0:
        print( f" {col:12s}:  {u} ({u/len(df)*100:.1f}%)")
```

<h2>12.Encoding Month Column</h2>
<p> Encoding Month Feature </p>
<p>The 'month' column is converted into numerical foramt:</p>
<p>Example: </p>
<p>-jan -> 1</p>
<p>-Feb -> 2</p>
<p> This makes it usable for machine learning algorithms</p>


```python
month_order = {'jan':1 ,'feb':2 ,'mar':3 ,'apr':4,'may':5,'jun':6,
               'jul':7 ,'aug':8 , 'sep':9,'oct':10, 'nov':11,'dec':12}
df['month_num'] = df['month'].map(month_order)

df['age_group']= pd.cut(df['age'],
                        bins=[17,29,39,49,59,100],
                        labels=['18-29','30-39','40-49','50-59','60+'])

df['balance_tier']= pd.cut(df['balance'],
                           bins=[-np.inf,0,500,2000,np.inf],
                           labels=['Negative','Low','Medium','High'])

print('New features added: month_num , age_group, balance_tier, was_previously_contacted')
print(df[['age_group','balance_tier','month_num', 'was_previously_contacted']].head())
```
